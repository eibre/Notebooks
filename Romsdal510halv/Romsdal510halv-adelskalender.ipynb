```python
import pandas as pd
import numpy as np
import os
pd.set_option('display.max_rows', 100)
path = r".\resultater"
filenames = os.listdir(path)
```

```python
df = None
for filename in filenames:
    filepath = path+'\\'+filename
    date = filename.split('.')[0]
    df_temp = pd.read_csv(filepath, sep='\t')
    df_temp['Date']=date
    if df is None:
        df = df_temp
    else:
        df = pd.concat([df_temp,df], ignore_index=True)
df = df.drop(columns=['Bib', 'First name', 'Last name', 'Team name 2', 'Start', 'Finish', 'Year of birth', 'Lap 1', 'Lap 2', 'Split 1', 'Split 2'])
```

## De 20 raskeste raskeste gjennom tidene:

```python
print("De 20 raskeste tidene på halvmaraton")
df[df['Distance'] == 'Halvmaraton'].sort_values('Time').reset_index(drop=True).shift()[1:].head(20)
```

```python
print('De 20 raskeste tidene på 10 km')
df[df['Distance'].str.contains('10 km')].sort_values('Time').reset_index(drop=True).shift()[1:].head(20)
```

```python
print("De 20 raskeste tidene på 5 km")
df[df['Distance'].str.contains('5 km')].sort_values('Time').reset_index(drop=True).shift()[1:].head(20)
```

## Deltakerantall

```python
print("Antall unike deltakere: " + str(len(pd.unique(df['Name']))))
print("Antall løp: " + str(len(filenames)))
```

```python
print("Mest populær distanse:")
df.groupby('Distance')['Distance'].count().sort_values(ascending=False)
```

```python
print('Ivrigste deltakere')
df.groupby('Name')['Name'].count().sort_values(ascending=False).head(20)
```

```python
print("Utvikling i deltakerantall (påmeldte)")
df.groupby('Date')['Date'].count()
```

```python
print("Antall fullførte")
df[~df['Time'].str.startswith('DN')].groupby('Date')['Date'].count()
```
